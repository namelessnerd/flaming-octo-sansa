```python
def permute_helper(input_string, current_permuted_string, permutations):
    input_length = len(input_string)
    if input_length == 0:
        print ("Permutation is " + current_permuted_string)
        permutations.append(current_permuted_string)
    set_of_remaining_strings_for_next_level = set()
    for pos in range(input_length):
        next_permutation = current_permuted_string + input_string[pos]
        remaining_string_to_permute = input_string[:pos] + input_string[pos + 1:]
        if not remaining_string_to_permute in set_of_remaining_strings_for_next_level:
            set_of_remaining_strings_for_next_level.add(remaining_string_to_permute)
            permute_helper(remaining_string_to_permute, next_permutation, permutations)

def permute(input_string):
    if input_string is not None and type(input_string) is str and len(input_string.strip()):
        permutations = []
        permute_helper(input_string, "", permutations)
        return permutations
    else:
        print ("Input to permute is a non empty string after it is stripped of trailing spaces")
        return False

def generate_palindrome(input_string):
    if input_string is not None and type(input_string) is str:
        input_string = input_string.strip()
        if input_string:
            # character_count_map - a map that stores how many times each character appears in the input string
            character_count_map = {}
            for ch in input_string:
                if ch in character_count_map:
                    character_count_map[ch] += 1
                else:
                    character_count_map[ch] = 1
            # palindrome_base (str) - the minimal set of characters in the input string that need to be present on 
            # one half of the palindrome
            palindrome_base = ""
            # odd_char (str) - in an odd length string, there can be one character that can be in the middle
            odd_char = ""
             # palindrome_base_odd (str) - in an odd length string, the character that appears odd number of times
             # will have one occuerence in the middle and the half of the rest on either side
            odd_char_palindrome_base = ""
            # num_odd_characters (int) - an odd length string can have character whose count is odd. An even length string
            # cannot have any
            num_odd_characters = 0
            for ch in character_count_map:
                # if the count is odd, % 2 will return 1. The if will eval to True
                if character_count_map[ch] % 2:
                    # reduce 1 from the number of odd characters. if there are more than one this value will be less 
                    # than 0
                    num_odd_characters += 1
                    # an char that appears odd number of times, one occurence has to be in the middle. 
                    # of the remaining, one half will be on each side of the palindrome. 
                    odd_char_palindrome_base += ch * ((character_count_map[ch]) //2)
                    odd_char = ch
                else:
                    palindrome_base += ch * ((character_count_map[ch]) // 2)
                    
            if len(input_string) % 2:
                # an odd length string if there are more than 1 odd characters, we reduce the count by 1. 
                # if there is one odd characters, num_odd_characters will be 0
                num_odd_characters -= 1
                
            if not num_odd_characters:
                # either there is one odd character or none
                palindrome_base = palindrome_base + odd_char_palindrome_base
                # permuted_base (list): generate all possible permutations of the base
                permuted_base = permute(palindrome_base)
                generated_palindromes = []
                for permuted in permuted_base:
                    generated_palindromes.append(permuted + odd_char + permuted[::-1])
                return generated_palindromes
            else:
                print ("Palindromes are not possible for %s" % input_string)
        else:
            return [""]
        
def is_palindrome_helper(input_string, left_index, right_index):
    if left_index > right_index:
        return True
    if input_string[left_index] == input_string[right_index]:
        return is_palindrome_helper(input_string, left_index + 1, right_index -1)
    return False

def is_palindrome(input_string):
    if input_string is not None and type(input_string) is str:
        input_string = input_string.strip().lower()
        if input_string:
            return is_palindrome_helper(input_string, 0, len(input_string) - 1)
                    
assert(is_palindrome(None) is None)
assert(is_palindrome(123) is None)
assert(is_palindrome([]) is None)
assert(is_palindrome("akKa"))
assert (list(filter (lambda x: is_palindrome(x) is False, generate_palindrome("a"*2 + "b"*2 + "C"*3))) == [])
assert(generate_palindrome("aaab") is None)

```
